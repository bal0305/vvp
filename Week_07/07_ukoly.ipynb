```python
import matplotlib.pyplot as plt 
import numpy as np
import math
x = np.linspace(-6.28, 6.28, 100)
fig = plt.figure(figsize=(10, 5))
plt.plot(x, np.sin(x), color='red', linewidth=3)
plt.xlabel('čas [s]')
plt.ylabel('amplituda [m]')
plt.title('Amplitudový průběh funkce závaží na pružině')
plt.show()
plt.savefig('graf.pdf', bbox_inches='tight')
```

1. Vykreslete mřížku n x n grafů (můžete začít s n=3).
- vyrobte list s hodnotami $\sin (ix)$ pro $i \in \langle 1, n \rangle$ pro $x \in \langle 0, 2\pi \rangle$
- do každého grafu (se souřadnicemi i - index řádku, j - index sloupce mřížky) vykreslete graf s x hodnotami dle i-tého prvku listu a s y hodnotami dle j-tého prvku listu

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(12,12))
for i in range(3):
    for j in range(3):
        x = np.linspace(0, 6.28, 100)
        y = np.sin(x*(i*3 + j))
        axes[i,j].plot(x, y)
        axes[i,j].set_title(f'Funkce sin(x*{i*3 + j})')
plt.tight_layout()
```
